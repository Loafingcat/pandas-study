```python
import pandas as pd

TITANIC_URL = (
  'https://raw.githubusercontent.com/'
  'datasciencedojo/datasets/master/titanic.csv'
)
titanic = pd.read_csv(TITANIC_URL)
```

```python
age_median = titanic['Age'].median()
titanic['Age'] = titanic['Age'].fillna(age_median)

print(f"나이(Age)의 중앙값: {age_median}세")
print(f"중앙값 대체 후 결측치 개수: {titanic['Age'].isnull().sum()}개\n")
```

```python
survived_only = titanic[titanic['Survived'] == 1]

print(f"생존자는 총 {len(survived_only)}명입니다.\n")
```

```python
sex_survival_rate = titanic.groupby('Sex')['Survived'].mean()

print("성별에 따른 생존율 결과:")
print(f"여성(female) 생존율: {sex_survival_rate['female'] * 100:.2f}%")
print(f"남성(male) 생존율: {sex_survival_rate['male'] * 100:.2f}%")
```

```python
titanic.groupby('Pclass').agg(
    생존율=('Survived', 'mean'),
    평균요금=('Fare',    'mean'),
    인원수 =('Survived', 'count')
)
```

```python
first_class = titanic[(titanic['Pclass'] == 1) & (titanic['Sex'] == 'female')]

print(first_class[['Name', 'Age']])
```

```python
Embarked =  titanic.groupby('Embarked')['Survived'].mean()

print("탑승 항구(Embarked)에 따른 생존율 결과:")
for port in Embarked.index:
    print(f"{port} 항구 생존율: {Embarked[port] * 100:.2f}%")
```

```python
import numpy as np
np.random.seed(42)
n = 150
shop = pd.DataFrame({
    'customer': np.random.choice(['김민준','이서현','박지우','최하은','정도윤'], n),
    'category': np.random.choice(['의류','전자기기','식품','생활용품','뷰티'], n),
    'quantity': np.random.randint(1, 6, n),
    'price':    np.random.choice([15000,29000,45000,89000,120000], n),
    'rating':   np.random.choice([1,2,3,4,5], n, p=[0.05,0.10,0.20,0.35,0.30]),
    'returned': np.random.choice([True, False], n, p=[0.1, 0.9])
})
shop.loc[np.random.choice(shop.index, 8, replace=False), 'rating'] = float('nan')
shop.to_csv('shop_orders.csv', index=False)

df = pd.read_csv('shop_orders.csv')
# Q1. 카테고리별 총 매출 내림차순
# Q2. 반품률 가장 높은 카테고리
# Q3. rating 결측치를 category별 평균으로 채우기
# Q4. 고객별 총 주문 금액 TOP 3
```

```python
# Q1. 카테고리별 총 매출 내림차순
category_sales = df.groupby('category').apply(lambda x: (x['quantity'] * x['price']).sum())
category_sales = category_sales.sort_values(ascending=False)
print("카테고리별 총 매출 (내림차순):")
print(category_sales)
```
